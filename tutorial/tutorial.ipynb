### 2. Loading Data

In the pickle file provided in the Github repo, we have a preprocessed PBMC dataset provided by 10x Genomics. We load that here using simple pandas functions:

```python
data = pd.read_pickle("/home/mkuchroo/temp_data/pbmc_4k_sqrt.pickle")
```

Next, we'll run the PHATE (https://www.nature.com/articles/s41587-019-0336-3), a dimensionality reduction algorithm previously developed by the Krishnaswamy lab to produce powerful visualizations.

```python
phate_op = phate.PHATE()
data_phate = phate_op.fit_transform(data)

scprep.plot.scatter2d(
    data_phate,
    fontsize=16,
    s=50,
    legend=False,
    ticks=False,
    label_prefix="PHATE",
    figsize=(8, 6),
)
```

### 3. Running CATCH on our dataset

First, we run the CATCH condensation process on the data as follows:

1. Compute the CATCH operator on our dataset
2. Identify clustering granularities amenable for hierarchical/topological analysis
3. Visualize our datasets with the chosen levels of resolution

```python
catch_op = catch.CATCH(knn=15, random_state=18, n_pca=100, n_jobs=1)
catch_op.fit(data)
```

Next, we compute granularities of clustering on this data 

```python
levels = catch_op.transform()
```

We can now identify granularities identified by topological activity analysis which partition single cells into meaningul, stable clusters:

```python
print(levels)

plt.plot(catch_op.gradient)
plt.xlabel("Iterations")
plt.ylabel("Topological Activity")
plt.scatter(len(catch_op.NxTs)+levels, catch_op.gradient[levels+1], c='r')

plt.show()
```

Finally, we can visualize the CATCH computed clusters on our initial PHATE embedding:

```python
fig, axes = plt.subplots(2,2, figsize=(10, 8))

for i, ax in enumerate(axes.flatten()):    
    scprep.plot.scatter2d(data_phate, c=catch_op.NxTs[levels[-i-1]], legend_anchor=(1,1), ax=ax,
                          title='Granularity '+str(len(catch_op.NxTs)+levels[-i-1]),
                          xticks=False, yticks=False, label_prefix="PHATE", fontsize=10, s=3)

fig.tight_layout()
```
