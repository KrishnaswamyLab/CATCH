```python
import numpy as np
import graphtools
import tasklogger
import collections
from collections import defaultdict
from scipy.spatial.distance import pdist, cdist, squareform
import sklearn
#import vne
import warnings
import phate
import CATCH

warnings.simplefilter("ignore")

import pandas as pd
import numpy as np
#%load_ext nb_black
```

```python
data = pd.read_pickle("/home/mkuchroo/temp_data/pbmc_4k_sqrt.pickle")
```

```python
phate_op = phate.PHATE()
data_phate = phate_op.fit_transform(data)
```

```python
from CATCH import catch
```

```python
### Running CATCH
```

```python
catch_op = catch.CATCH(knn=10, random_state=1, n_pca=25, n_jobs=5)
```

```python
catch_op.fit(data)
```

```python
levels = catch_op.transform()
```

```python
levels
```

```python
len(catch_op.NxTs)
```

```python
import matplotlib.pyplot as plt

plt.plot(catch_op.gradient)
```

```python
import scprep

scprep.plot.scatter2d(
    data_phate,
    c=catch_op.NxTs[levels[-4]],
    fontsize=16,
    s=50,
    legend=False,
    ticks=False,
    label_prefix="PHATE",
    figsize=(8, 6),
)
```
